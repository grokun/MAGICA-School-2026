## Lecture: CF and UGRID Standards

CF (Climate and Forecast) Metadata Conventions standardize how variables, coordinates, and units are described in NetCDF and Zarr datasets, enabling tools to work across models and instruments without knowing dataset-specific variable names. UGRID extends CF for unstructured meshes (triangles, polygons) used by coastal and ocean models like FVCOM, SCHISM, and SHYFEM.
